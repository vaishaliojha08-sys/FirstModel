```python
#Using requests fetch multiple pages from Books to Scrape (simulate 3 pages)
#Parse titles from all pages using BeautifulSoup and store in a list

base_url = "https://books.toscrape.com/index.html"
books_info = requests.get(base_url).content
book_titles = []
for page_num in range(1,4):
  page_url =f"https://books.toscrape.com/catalogue/page-{page_num}.html"
  print(f"Fetching URL {page_url}")
  page = requests.get(page_url)
  soup = BeautifulSoup(page.text, "html.parser")
  books = soup.findAll(name="li",class_="col-xs-6 col-sm-4 col-md-3 col-lg-3")
  #books = soup.findAll("article", class_="product_pod")
  for book in (books):
    book_title = book.h3.a["title"]
    book_titles.append(book_title)

print(*book_titles, sep='\n')
```
