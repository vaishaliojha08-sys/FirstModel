```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# -----------------------------
# Step 1: Recreate the dataset
# -----------------------------

# Confusion matrix values
TN = 4600
FP = 150
FN = 100
TP = 150

# Create actual labels
y_true = np.array(
    [0]*TN + [0]*FP +   # actual no default
    [1]*FN + [1]*TP     # actual default
)

print(y_true)

# Create predicted labels
y_pred = np.array(
    [0]*TN + [1]*FP +   # predictions for non-defaults
    [0]*FN + [1]*TP     # predictions for defaults
)


# -----------------------------
# Step 2: Calculate metrics
# -----------------------------

accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)

print("Model Performance Metrics")
print("-------------------------")
print("Accuracy:", round(accuracy,4))
print("Precision:", round(precision,4))
print("Recall:", round(recall,4))
print("F1 Score:", round(f1,4))

# -----------------------------------------------------------
# Explanation 1: Why accuracy alone is misleading
# -----------------------------------------------------------

"""
Accuracy = (TP + TN) / Total

In this dataset:
Total applications = 5000
Actual defaults = 250
Non-defaults = 4750

This means the dataset is highly imbalanced because only 5% of the applicants default.

A naive model could predict "No Default" for every applicant and still achieve:
Accuracy = 4750 / 5000 = 95%

However, such a model would completely fail to identify any defaulters.

Therefore, accuracy is misleading because it hides the model's ability to detect
the minority but critical class (loan defaulters). Metrics like precision,
recall, and F1-score are better indicators for this problem.
"""

# -----------------------------
# Step 3: Display Confusion Matrix
# -----------------------------

cm = confusion_matrix(y_true, y_pred)
print("\nConfusion Matrix:")
print(cm)

# -----------------------------
# Step 4: Business Cost Analysis
# -----------------------------

cost_FN = 500000
cost_FP = 5000

total_FN_cost = FN * cost_FN
total_FP_cost = FP * cost_FP

total_cost = total_FN_cost + total_FP_cost

print("\nBusiness Cost Analysis")
print("----------------------")
print("False Negative Cost:", total_FN_cost)
print("False Positive Cost:", total_FP_cost)
print("Total Cost:", total_cost)

# -----------------------------
# Step 5: Put results into dataframe
# -----------------------------

results = pd.DataFrame({
    "Metric":["Accuracy","Precision","Recall","F1 Score"],
    "Value":[accuracy,precision,recall,f1]
})

print("\nMetrics Table")
print(results)

# -----------------------------------------------------------
# Explanation 2: Which metric should be prioritized
# -----------------------------------------------------------

"""
False Negative Cost = Rs. 5,00,000
False Positive Cost = Rs. 5,000

Total FN Cost = 100 * 5,00,000 = Rs. 5,00,00,000
Total FP Cost = 150 * 5,000 = Rs. 7,50,000

The cost of missing a defaulter (FN) is dramatically higher than the cost of
reviewing a safe applicant (FP).

Therefore, the company should prioritize RECALL.

Recall measures how many actual defaulters the model correctly identifies.

Higher recall reduces False Negatives, which directly reduces the extremely
high financial loss from undetected defaulters.


Threshold Adjustment:

Currently the model uses a probability threshold of 0.5.

If we LOWER the threshold (e.g., to 0.3):
- More applicants will be predicted as "Default"
- False Negatives will decrease
- Recall will increase
- False Positives may increase

Even if FP increases slightly, the extra manual review cost is small compared
to the massive loss from missing a real defaulter.

Therefore, adjusting the classification threshold to increase recall can
significantly reduce the company's total financial losses.
"""
```
