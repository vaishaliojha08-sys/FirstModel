```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix

# ===============================
# Business Parameters
# ===============================

revenue_per_purchase = 50
cost_per_recommendation = 0.10

# ===============================
# Given Confusion Matrix Values
# ===============================

# Threshold = 0.5
tp_05 = 150
fp_05 = 1500
fn_05 = 50
tn_05 = 8300

# Threshold = 0.3
tp_03 = 180
fp_03 = 3200
fn_03 = 20
tn_03 = 6600


# ===============================
# Profit Calculation Function
# ===============================

def calculate_profit(tp, fp, revenue_per_purchase, cost_per_recommendation):

    revenue = tp * revenue_per_purchase
    recommendations = tp + fp
    cost = recommendations * cost_per_recommendation
    profit = revenue - cost

    return revenue, cost, profit


# ===============================
# Threshold 0.5 Calculations
# ===============================

rev_05, cost_05, profit_05 = calculate_profit(
    tp_05, fp_05, revenue_per_purchase, cost_per_recommendation
)

# ===============================
# Threshold 0.3 Calculations
# ===============================

rev_03, cost_03, profit_03 = calculate_profit(
    tp_03, fp_03, revenue_per_purchase, cost_per_recommendation
)


# ===============================
# Create Comparison Table
# ===============================

results = pd.DataFrame({
    "Threshold": [0.5, 0.3],
    "True Positives": [tp_05, tp_03],
    "False Positives": [fp_05, fp_03],
    "Revenue ($)": [rev_05, rev_03],
    "Cost ($)": [cost_05, cost_03],
    "Net Profit ($)": [profit_05, profit_03]
})

print("\nBusiness Profit Comparison:\n")
print(results)


# ===============================
# Recommendation
# ===============================

if profit_03 > profit_05:
    print("\nRecommendation: Switch to threshold 0.3 because it generates higher profit.")
else:
    print("\nRecommendation: Keep threshold 0.5 because it generates higher profit.")
```
